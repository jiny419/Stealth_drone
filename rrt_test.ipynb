```python
import rrt
```

```python
X_start = rrt.SE2_from_param([0, 1, 0])  # theta=0, x=0, y=0
X_goal = rrt.SE2_from_param([0, 8, 0])  # theta=0, x=10, y=0

# this is a list of all obstacles
# x, y, radius
collision_points = [
    [6, 0, 1],
    [2, 4, 1],
    [5, -5, 2],
    [10, 5, 3],
    [10, -5, 1],
]

ret = rrt.rrt(X_start=X_start, X_goal=X_goal, vehicle_radius=1,
        box=[0, 10, -5, 5], collision_points=collision_points,
        plot=True, max_iterations=700, dist_plan=3, tolerance=0.5)
print('success', ret['success'])
```
